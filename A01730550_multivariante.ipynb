```python
import pandas as pd
import numpy as np # for getting our distribution
import matplotlib.pyplot as plt # for plotting
import seaborn as sns; sns.set()
df = pd.read_csv('datasets/antropometria-dataset.csv')
pd.DataFrame.hist(data=df,
                 column=['sistol','diastol'])
plt.show()
```

```python
pd.DataFrame.hist(data=df,
                 column=['peso','cintura'])
plt.show()
```

```python
df.plot.scatter(x='peso',
               y='cintura')
plt.show()
```

```python
sexo1= df['sexo']==1 #cambiar a booleano
df1=df[sexo1] 
pd.DataFrame.hist(data=df1,
                 column=['sistol','diastol','peso','cintura'])
plt.show()
```

```python
sexo2= df['sexo']==2
df2=df[sexo2]
pd.DataFrame.hist(data=df2,
                 column=['sistol','diastol','peso','cintura'])
plt.show()
```
