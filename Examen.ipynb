```python
pd.pivot_table(df,
              index =["month","day"],
              values = "temp",
              aggfunc="mean")
```

```python
def create_category(ranking):
    if (ranking>=0) &(ranking<=10):
        return "Dias frios"
    elif (ranking >10) &(ranking<=20):
        return "Dias normales"
    elif (ranking>20) &(ranking<=30):
        return "Dias calurosos"
    else:
        "Dias super calurosos "
        
```

```python
df["rank"]=df["temp"].apply(lambda x:create_category(x))
df
```
