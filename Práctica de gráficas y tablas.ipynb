```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
r=1
mean=[15,5]
cov=[[1,r],[r,1]]
x,y= x,y = np.random.multivariate_normal(mean,cov,400).T
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist(x=x,
        bins=10
        )
plt.subplot(1,2,2)
plt.hist(x=y,
        bins=10)
plt.title('Y')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.scatter(x=x,
           y=y)
plt.title('Joint distribution of x and y')
plt.subplot(222)
plt.hist(x=x,
        bins=10)
plt.title('Marginal distribution of X')

plt.subplot(224)
plt.hist(x=y,
        orientation = 'horizontal',
        bins = 15)
plt.title('Marginal distribution of y')
plt.show()

#En el archivo 7 03 hay más gráficos que hay que practicar de este mismo estilo
```

```python
import numpy as np
import pandas as pd
df = pd.read_excel('datasets/Canada.xlsx',
                  sheet_name='Canada by Citizenship',
                  skiprows = range(20),
                  skipfooter = 2,
                  engine='openpyxl'
                  )
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['AREA','REG','DEV', 'Type','Coverage'],
        axis=1,
        inplace=True)
df.head()
```

```python
df.rename(columns ={'OdName':'Contry',
                   'AreaName': 'Continent',
                   'RegName': 'Region'
                   },
         inplace= True)
df.head()
```

```python
df.set_index('Contry', inplace = True)
```

```python
df.head()
```

```python
df['Total']=df.iloc[:, 4:].sum(axis=1)
df.head()
```

```python
years = list (map(str, range (1980, 2014)))
```

```python
years
```

```python
import matplotlib as mpl
mpl.style.use('ggplot')
```

```python
df_continents = df.groupby('Continent',
                          axis=0).sum()
df_continents
```

```python
#color_list = ['gold', 'yellowgreen', 'lightskyblue', 'lightgreen', 'pink']
#explode_list=[0.1, 0, 0, 0, 0.2, 0.3]
df_continents['Total'].plot(kind = 'pie',
                           figsize =(5,8),
                           autopct = '%10.1f%%',
                           startangle= 200,
                           shadow =True
                           )
plt.show()
```
