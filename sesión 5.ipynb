```python
path="datasets/Cartwheeldata.csv"
df=pd.read_csv(path)  #si ya nombraste pandas arriba, aqui ya no lo vuelves a nombrar, te vas directo
df.head(2)
```

```python
import seaborn as sns #Esto es para que te salgan graficos estadisticos boxplot,regresión
sns.lmplot(x="Wingspan",
          y="CWDistance",
          data=df,
          fit_reg= True,#Si es true, te sale linea de regresión, si es false se quita tu linea de regresion 
          hue="Gender")  # color dependiendo a la columna 
plt.title("Scatter plot of wingspan vs CWdistance ")
plt.show()
```

```python
sns.swarmplot(x="Gender",
            y="CWDistance",
            data=df)
plt.title("Gender vs CWDistance")
plt.show()
```

```python
sns.boxplot(data=df.loc[df["Gender"]=='M',
            ['Age','Height','Wingspam']])
plt.show()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(1,2,1) # 1 renglon 2 columnas,  pos 1
sns.boxplot(data=df.loc[df['Gender']=='M',
                       ['Score']
                       ]
           )

ax2=fig.add_subplot(1,3,3)
sns.boxplot(data=df.loc[df['Gender']=='F',
                       ['Score']
                       ]
           )

ax1.title.set_text("Male score")
ax2.title.set_text("Female score")
plt.show()
```

```python
import seaborn as sns
sns.distplot(df.DWDdistance)
plt.show()
```
